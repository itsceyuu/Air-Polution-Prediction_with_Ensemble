```python
from pathlib import Path

def inspect_dataframe(df, file_name):
    html = f"<h2>File: {file_name}</h2>"

    # Shape
    html += f"<p><b>Shape:</b> {df.shape}</p>"

    # Columns
    html += "<h3>Columns</h3>"
    html += pd.DataFrame(df.columns, columns=["columns"]).to_html(index=False)

    # Dtypes
    html += "<h3>Data Types</h3>"
    html += df.dtypes.to_frame("dtype").to_html()

    # Missing values
    html += "<h3>Missing Values</h3>"
    html += df.isna().sum().to_frame("missing_count").to_html()

    # Head & Tail
    html += "<h3>Head</h3>"
    html += df.head(3).to_html()

    html += "<h3>Tail</h3>"
    html += df.tail(3).to_html()

    # Describe numeric
    html += "<h3>Describe (Numeric)</h3>"
    html += df.describe().to_html()

    # Sample tanggal kalau ada
    for col in df.columns:
        if "tanggal" in col.lower() or "time" in col.lower():
            html += f"<h3>Sample Date Column: {col}</h3>"
            html += df[col].head(10).to_frame().to_html()
            break

    # Unique important columns
    important_cols = ["stasiun", "kategori", "parameter_pencemar_kritis"]
    for col in important_cols:
        if col in df.columns:
            html += f"<h3>Unique values: {col}</h3>"
            html += pd.DataFrame(df[col].unique(), columns=[col]).head(20).to_html(index=False)

    html += "<hr>"
    return html

def read_file(path):
    try:
        if path.endswith(".csv"):
            return pd.read_csv(path)
        elif path.endswith(".xlsx"):
            return pd.read_excel(path)
    except Exception as e:
        print(f"Gagal baca {path}: {e}")
        return None


```

```python
ALL_DIRS = [
    ISPU_DIR,
    CUACA_DIR,
    JP_DIR,
    KAS_DIR,
    LN_DIR,
    NDVI_DIR
]

report_html = "<h1>DATASET INSPECTION REPORT</h1>"

for folder in ALL_DIRS:
    folder_path = Path(folder)
    
    report_html += f"<h1>Folder: {folder}</h1>"
    
    for file in os.listdir(folder):
        file_path = folder_path / file
        
        df = read_file(str(file_path))
        if df is not None:
            report_html += inspect_dataframe(df, file)

with open("dataset_report.html", "w", encoding="utf-8") as f:
    f.write(report_html)

print("Report berhasil dibuat: dataset_report.html")
```
