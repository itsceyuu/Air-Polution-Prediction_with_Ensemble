## Trying Minimodel 

```python
# Load and inspect the provided preprocessed ISPU dataset
import pandas as pd

df = pd.read_csv("./datasets/ispu_preprocessed_4.csv")

# Basic inspection outputs
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nInfo:")
print(df.info())

print("\nMissing values per column:")
print(df.isna().sum().sort_values(ascending=False).head(20))

print("\nSample rows:")
print(df.head())

df = df.rename(columns={
    "pm_sepuluh": "pm10",
    "pm_duakomalima": "pm25",
    "sulfur_dioksida": "so2",
    "karbon_monoksida": "co",
    "ozon": "o3",
    "nitrogen_dioksida": "no2"
})

pollutant_cols = [
    "pm10",
    "pm25",
    "so2",
    "co",
    "o3",
    "no2"
]

print("\nMissing values (pollutants):")
print(df[pollutant_cols].isna().sum())

print("\nBasic statistics (pollutants):")
print(df[pollutant_cols].describe())

# Check unique categories and stations
print("\nUnique kategori:", df['kategori'].unique())
print("Unique stasiun:", df['stasiun'].unique())
```

```python
df_known = df[df["kategori"] != "TIDAK ADA DATA"].copy()
df_unknown = df[df["kategori"] == "TIDAK ADA DATA"].copy()
```

```python
df_known
```

```python
df_unknown
```

```python

drop_cols = ["tanggal", "kategori", "periode_data"]

features = [c for c in df_known.columns if c not in drop_cols]

X_train = df_known[features]
y_train = df_known["kategori"]

cat_cols = ["stasiun", "parameter_pencemar_kritis"]

X_unknown = df_unknown[features].copy()

for c in cat_cols:
    X_unknown[c] = X_unknown[c].astype(str).fillna("UNKNOWN")

for c in cat_cols:
    X_train[c] = X_train[c].astype(str).fillna("UNKNOWN")
```

```python
from catboost import CatBoostClassifier

model_imp = CatBoostClassifier(
    iterations=500,
    depth=6,
    learning_rate=0.05,
    verbose=100
)

model_imp.fit(X_train, y_train, cat_features=cat_cols)
```

```python
# X_unknown = df_unknown[features]
# pred_kategori = model_imp.predict(X_unknown)
pred_kategori = model_imp.predict(
    X_unknown,
    prediction_type='Class'
).ravel()      # <-- ini yang bikin 1D
df_unknown["kategori"] = pred_kategori

df_filled = pd.concat([df_known, df_unknown]).sort_values("tanggal")
```

```python
model_imp.save_model("catboost_minimodel_2.cbm")
```

## Trying XGB 

```python
# Load and inspect the provided preprocessed ISPU dataset
import pandas as pd

df_xgb = pd.read_csv("./datasets/ispu_imputed_fixed.csv")

# Basic inspection outputs
print("Shape:", df_xgb.shape)
print("\nColumns:")
print(df_xgb.columns.tolist())

print("\nInfo:")
print(df_xgb.info())

print("\nMissing values per column:")
print(df_xgb.isna().sum().sort_values(ascending=False).head(20))

print("\nSample rows:")
print(df_xgb.head())

df_xgb = df_xgb.rename(columns={
    "pm_sepuluh": "pm10",
    "pm_duakomalima": "pm25",
    "sulfur_dioksida": "so2",
    "karbon_monoksida": "co",
    "ozon": "o3",
    "nitrogen_dioksida": "no2"
})

pollutant_cols = [
    "pm10",
    "pm25",
    "so2",
    "co",
    "o3",
    "no2"
]

print("\nMissing values (pollutants):")
print(df_xgb[pollutant_cols].isna().sum())

print("\nBasic statistics (pollutants):")
print(df_xgb[pollutant_cols].describe())

# Check unique categories and stations
print("\nUnique kategori:", df_xgb['kategori'].unique())
print("Unique stasiun:", df_xgb['stasiun'].unique())
```

```python
df_xgb["tanggal"] = pd.to_datetime(df_xgb["tanggal"])
split_date = df_xgb["tanggal"].quantile(0.8)

train = df_xgb[df_xgb["tanggal"] <= split_date]
test  = df_xgb[df_xgb["tanggal"] > split_date]

drop_cols = [
    "tanggal",
    "kategori",
    "parameter_pencemar_kritis",
    "periode_data", 
    "max"
]

features = [c for c in df_xgb.columns if c not in drop_cols]

X_train = train[features]
y_train = train["kategori"]

X_test = test[features]
y_test = test["kategori"]
```

```python
# xgb buat yg 5 kategori dgn berbahaya 
from sklearn.preprocessing import LabelEncoder

le_station = LabelEncoder()
X_train["stasiun"] = le_station.fit_transform(X_train["stasiun"])
X_test["stasiun"] = le_station.transform(X_test["stasiun"])

le_target = LabelEncoder()
y_train_enc = le_target.fit_transform(y_train)
y_test_enc = le_target.transform(y_test)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=1000,
    max_depth=6,
    learning_rate=0.05,
    objective='multi:softmax',
    num_class=3,
    eval_metric='mlogloss',
    use_label_encoder=False,
    verbosity=1
)

model.fit(X_train, y_train_enc)

from sklearn.metrics import classification_report

y_pred_enc = model.predict(X_test)
y_pred = le_target.inverse_transform(y_pred_enc)

print(classification_report(y_test, y_pred))
```
